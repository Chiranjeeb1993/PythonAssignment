```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
prd_url=[]

def scrape_product_page(url):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
#     products = soup.find_all('div', {'data-component-type': 's-search-result'})

    page = soup.find_all('div', {'data-component-type': 's-search-result'})
    
#     link = page.find('a', {'class': 'a-link-normal s-no-outline'})


    for pages in page:
        
        url = "https://www.amazon.in" + pages.find('a', {'class': 'a-link-normal s-no-outline'})['href']
        prd_url.append(url)
        act_url=prd_url[-1]
        print(act_url)
    
    
```

```python
def scrape_multiple_pages(base_url, num_pages):
    
    for page in range(1, num_pages + 1):
        url = f"{base_url}&page={page}"
        scrape_product_page(url)
#         scrape_all_product(url)


base_url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
num_pages = 20

scrape_multiple_pages(base_url, num_pages)
```

```python
def scrape_all_product(url):
    all_prd_url=[]
    
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    page = soup.find_all('div', {'data-component-type': 's-search-result'})


    for pages in page:
        
        url = "https://www.amazon.in" + pages.find('a', {'class': 'a-link-normal s-no-outline'})['href']
        
        product_name = pages.find('span', {'class': 'a-size-medium a-color-base a-text-normal'}).text
        product_price = pages.find('span', {'class': 'a-offscreen'})
        rating = pages.find('span', {'class': 'a-icon-alt'}).text
        num_reviews = pages.find('span', {'class': 'a-size-base'}).text
        
        all_prd_url.append({
        'Product URL': url,
        'Product Name': product_name,
        'Product Price': product_price.text,
        'Rating': rating,
        'Reviews': num_reviews
            })
        
        act_url=all_prd_url[-1]
        
        print(act_url)
        
#         data = pd.DataFrame(all_prd_url)
#         print(data)
        
    
```

```python
def scrape_multiple_pages(base_url, num_pages):
    
    for page in range(1, num_pages + 1):
        url = f"{base_url}&page={page}"
#         scrape_product_page(url)
        scrape_all_product(url)

base_url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
num_pages = 20

scrape_multiple_pages(base_url, num_pages)
```
